Part 8: Data Cleaning

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Marveen\Downloads\Customer Call List.xlsx")
df
```

REMOVING DUPLICATES

```python
df = df.drop_duplicates()
df
```

DROPPING UNNECESSARY COLUMNS

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df.info()
```

CLEANING THE DATA VALUES

```python
df['Last_Name'] = df['Last_Name'].str.strip(".../_123")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex = True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + "-" + x[3:6] + "-" + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--","")
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--","")
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', n=2, expand = True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].replace({"N":"No","Y":"Yes"})
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("N/a","")
df
```

```python
df.fillna("")
```

```python
df = df.replace("N/a","")
df = df.fillna("")
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"Y":"Yes"})
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"Yeses":"Yes"})
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"N":"No"})
df
```

```python
df = df.loc[df["Do_Not_Contact"] == "No"]
df
```
